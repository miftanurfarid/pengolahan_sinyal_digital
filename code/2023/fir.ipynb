```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
fs = 8000
b = np.array([0.1871, 0.2, 0.1871])
w, h = signal.freqz(b)
plt.figure(figsize=(10,3))
plt.title('Digital filter frequency response')
plt.plot(w*fs/(2*np.pi), 20 * np.log10(abs(h)), 'b')
plt.ylabel('Magnitude (dB)', color='b')
plt.xlabel('Frequency (Hz)')
plt.legend(['3-tap FIR'], loc='lower left')
plt.grid()
plt.show()
```

```python
angles = np.unwrap(np.angle(h))
plt.figure(figsize=(10,3))
plt.title('Digital filter frequency response')
plt.plot(w*fs/(2*np.pi), angles*180/np.pi, 'g')
plt.ylabel('Phase (Degree)', color='g')
plt.xlabel('Frequency (Hz)')
plt.legend(['3-tap FIR'], loc='lower left')
plt.grid()
plt.show()
```

```python
ntaps = 25
fs = 8000
f1 = 2000*2*np.pi/fs
```

```python
b_rec = signal.firwin(ntaps, f1, fs=8000*2*np.pi, window='boxcar')
b_rec
```

```python
b_ham = signal.firwin(ntaps, f1, fs=8000*2*np.pi, window='hamming')
b_ham
```

```python
w_rec, h_rec = signal.freqz(b_rec)
w_ham, h_ham = signal.freqz(b_ham)
```

```python
plt.figure(figsize=(10,2))
plt.title('Digital filter frequency response')
plt.plot(w_rec*fs/(2*np.pi), 20 * np.log10(abs(h_rec)), 'b', w_ham*fs/(2*np.pi), 20 * np.log10(abs(h_ham)), 'r')
plt.ylabel('Magnitude (dB)', color='b')
plt.xlabel('Frequency (Hz)')
plt.legend(['Rectangular','Hamming'], loc='lower left')
plt.grid()
plt.show()
```

```python
angles_rec = np.unwrap(np.angle(h_rec))
angles_ham = np.unwrap(np.angle(h_ham))
plt.figure(figsize=(10,2))
plt.title('Digital filter frequency response')
plt.plot(w_rec*fs/(2*np.pi), angles_rec*180/np.pi, 'b', w_ham*fs/(2*np.pi), angles_ham*180/np.pi, 'r')
plt.ylabel('Phase (Degree)', color='g')
plt.xlabel('Frequency (Hz)')
plt.legend(['Rectangular','Hamming'], loc='lower left')
plt.grid()
plt.show()
```

## Noise Reduction

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.fft import fft, fftfreq
```

```python
fs = 8000
```

```python
T = 1/fs
```

```python
v = np.sqrt(0.1)*np.random.randn(1,250)
v = v.T
```

```python
v.shape
```

```python
n = np.array([np.arange(0,250)])
n = n.T
```

```python
plt.plot(n,v)
plt.xlabel('Number of samples')
plt.ylabel('Sample value')
plt.legend(['broadband noise'])
plt.grid()
plt.show()
```

```python
x = np.sqrt(2)*np.sin(2*np.pi*500*n*T)
xv = x + v
```

```python
plt.plot(n,x)
plt.xlabel('Number of samples')
plt.ylabel('Sample value')
plt.legend(['signal'])
plt.grid()
plt.show()
```

```python
plt.plot(n,xv)
plt.xlabel('Number of samples')
plt.ylabel('Sample value')
plt.legend(['noisy signal'])
plt.grid()
plt.show()
```
